```python
import pandas as pd
import numpy as np
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
match.shape
```

```python
delivery.head()
```

```python
len(pd.unique(delivery['match_id'])) ## 756 matches were played in this time frame 
```

```python
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
total_score_df
```

```python
## now we only data for 1st innings as we will calculate or predict for 2nd innings 
total_score_df=total_score_df[total_score_df['inning']==1]
```

```python
# in macthes dataset we didn't had infirmation regrading how many runs were made in 1st innings now 
# that we have that info in total_score_df we will use this and merge with matches dataset

match_df = match.merge(total_score_df[['match_id','total_runs']],left_on = 'id',right_on = 'match_id')
```

```python
match_df.head()
```

```python
## we will only keep those teams in our data set which are currently playing ipl like pune warriors is not playing so we will not keep that
match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
#match_df['dl_applied'][match_df['dl_applied']==1]
#match_df['dl_applied'].value_counts()
match_df = match_df[match_df['dl_applied']==0]
```

```python
match_df.shape
```

```python
match_df
```

```python
## keeping only useful features
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
## now in delivery_df we only need value of innings =2 
delivery_df = delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.shape
```

```python
## now we have to calculate runs_left balls_left wickets_left curr run rate req run rate and result 
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score'] 
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
# crr = runs/over
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
## to calculate winner or loser we will comapre winner and batting team column if teams is both are same 
# then result will be 1 if different teams then 0
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets',
             'total_runs_x' , 'crr' , 'rrr' , 'result']]
```

```python
## now we will have to shuffle as all balls of one match are in order and this could create bias dataset
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left']!=0]
```

```python
### MACHINE LEARNING 
X = final_df.iloc[:,:-1]
Y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
## our three columns are in string (batting team , bowling team , city ) so we need to apply one_hot_encoding 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn import svm
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
    #('step2' ,svm.SVC())
    #('step2',RandomForestClassifier())
])
```

```python
pipe.fit(x_train,y_train)
```

```python
#x_train.describe()
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
